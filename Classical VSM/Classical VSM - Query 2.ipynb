# Essential Libraries

```python
import spacy
import neattext.functions as fn #libray for cleaning data
from sklearn.feature_extraction.text import CountVectorizer   #for Count vectorizing
from sklearn.feature_extraction.text import TfidfVectorizer #for tf-idf
import pandas as pd #for data handling
#libraries for Count Vectorization 
import re
import string
import nltk
```

# Loading the testbed dataset

```python
#load text file to work with
#article on ghana
gh_file = open('ghana.txt', 'r')
gh_file = gh_file.read()
gh_file = str(gh_file)
# print(gh_file)
# gh_file.close()

#article on us
us_file = open("wiki_us.txt", 'r')
us_file = us_file.read()
us_file = str(us_file)
# us_file.close()

#article on politics
politics_file = open('politics.txt', 'r')
politics_file = politics_file.read()
politics_file = str(politics_file)
# politics_file.close()

#article on Human Chimera
chimera_file = open('chimera.txt', 'r')
chimera_file = chimera_file.read()
chimera_file = str(chimera_file)

#article on computer science
computer_file = open("computer.txt", 'r', encoding = "utf-8")
computer_file = computer_file.read()
computer_file = str(computer_file)

#article on mathematics
math_file = open('math.txt', 'r', encoding = 'utf=8')
math_file = math_file.read()
math_file = str(math_file)

#article on enzyme
enzyme_file = open('enzyme.txt', 'r', encoding = 'utf-8')
enzyme_file = enzyme_file.read()
enzyme_file = str(enzyme_file)

#article on finance 
finance_file = open('finance.txt', 'r', encoding = 'utf-8')
finance_file = finance_file.read()
finance_file = str(finance_file)

#article on Ghana Bank crisis
gh_bank_file = open("gh_banking_crisis.txt", 'r', encoding = 'utf-8')
gh_bank_file = gh_bank_file.read()
gh_bank_file = str(gh_bank_file)

#article on Ghana economic turmoil
gh_economic_file = open("gh_economic_turmoil.txt", 'r', encoding = 'utf-8')
gh_economic_file = gh_economic_file.read()
gh_economic_file = str(gh_economic_file)

#article on World Financial crisis and its implicaitons on Ghana
gh_world_finance_file = open("gh_world_financial_crisis.txt", 'r', encoding = 'utf-8')
gh_world_finance_file = gh_world_finance_file.read()
gh_world_finance_file = str(gh_world_finance_file)

#article on the human body
human_body_file = open("human_body.txt", 'r', encoding = 'utf-8')
human_body_file = human_body_file.read()
human_body_file = str(human_body_file)

#query 
query = 'What are the functions of micro organisms in the stomach?'
```

# Using spcay to create document containers for further processing

```python
nlp = spacy.load("en_core_web_lg") #loading the spacy large model
```

```python
#creating containers for data files for easier manupulation
gh_doc  = nlp(gh_file)
us_doc = nlp(us_file )
politics_doc = nlp(politics_file)
chimera_doc = nlp(chimera_file)
computer_doc = nlp(computer_file)
math_doc = nlp(math_file)
enzyme_doc = nlp(enzyme_file)
finance_doc = nlp(finance_file)
gh_bank_doc = nlp(gh_bank_file)
gh_economic_doc = nlp(gh_economic_file)
gh_world_finance_doc = nlp(gh_world_finance_file)
human_body_doc = nlp(human_body_file)

query_doc = nlp(query)
```

```python
# Creating a corpus for the documents
corpus = [None,None,None,None,None, None, None, None, None, None, None, None, None]
corpus[0] = query_doc.text.lower()
corpus[1] = gh_doc.text.lower()
corpus[2] = us_doc.text.lower()
corpus[3] = politics_doc.text.lower()
corpus[4] = chimera_doc.text.lower()
corpus[5] = computer_doc.text.lower()
corpus[6] = math_doc.text.lower()
corpus[7] = enzyme_doc.text.lower()
corpus[8] = finance_doc.text.lower()
corpus[9] = gh_bank_doc.text.lower()
corpus[10] = gh_economic_doc.text.lower()
corpus[11] = gh_world_finance_doc.text.lower()
corpus[12] = human_body_doc.text.lower()

corpus
```

# TF-IDF

```python
tv = TfidfVectorizer()
tv = tv.fit(corpus)
corpus_vectors = tv.transform(corpus)
corpus_vectors = corpus_vectors.toarray()
# corpus_vector

df_tfidf = pd.DataFrame(corpus_vectors, columns = tv.get_feature_names())
df_tfidf = df_tfidf.transpose()
df_tfidf
```

```python
#renaming the columns to document names
tf = df_tfidf.rename(columns={0:"Query",1:"Ghana",2:"US",3:"Politics",4:"Chimera", 5:"Computer", 6:'math', 7:'enzyme',
                          8:'finance',
                         9:'Ghana Bank crisis',
                         10:'Ghana economic turmoil',
                         11:'world fin crisis-gh',
                         12:"Human body"})
tf.head(20)
```

# Cosine Similarity 

```python
#Cosine similarity for query and Ghana
from math import sqrt
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['Ghana']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_gh = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print("Cosine Ghana ->", cosine_gh)

#Cosine similarity for query and US
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['US']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_us = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print('Cosine US ->', cosine_us)


#Cosine similarity for query and Politics
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['Politics']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_Politics = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print('Cosine Politics ->', cosine_Politics)

#Cosine similarity for query and Chimera
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['Chimera']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_Chimera = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print('Cosine Chimera ->', cosine_Chimera)

#Cosine similarity for query and Computer
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['Computer']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_Computer = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print('Cosine Computer ->', cosine_Computer)

#Cosine similarity for query and math
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['math']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_math = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print('Cosine math ->', cosine_math)

#Cosine similarity for query and enzyme
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['enzyme']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_enzyme = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print('Cosine enzyme ->', cosine_enzyme)

#Cosine similarity for query and finance
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['finance']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_finance = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print('Cosine finance ->', cosine_finance)


#Cosine similarity for query and Ghana Bank crisis
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['Ghana Bank crisis']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_ghBank = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print('Cosine Ghana Bank crisis ->', cosine_ghBank)

#Cosine similarity for query and Ghana economic turmoil
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['Ghana economic turmoil']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_ghEcon = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print('Cosine Ghana economic turmoil ->', cosine_ghEcon)

#Cosine similarity for query and world fin crisis-gh
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['world fin crisis-gh']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_worldFin = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print('Cosine world fin crisis-gh ->', cosine_worldFin)


#Cosine similarity for query and Human body
dot_product = 0
sum_q = 0
sum_d = 0
for i,j in zip(tf['Query'],tf['Human body']):
    sum_q += i*i
    sum_d += j*j
    dot_product +=i*j
cosine_humanB = dot_product/((sqrt(sum_q)) * (sqrt(sum_d)))
# print('Cosine Human body ->', cosine_humanB)
```

```python
cosine_df = pd.DataFrame({'Ghana':[cosine_gh], 'US':[cosine_us], 
                    'Politics ': [cosine_Politics], 'Chimera':[cosine_Chimera], 
                    "Computer":[cosine_Computer], 'math': [cosine_math],
                    "enzyme":[cosine_enzyme],"finance":[cosine_finance],
                    "Ghana Bank crisis":[cosine_ghBank], "Ghana economic turmoil":[cosine_ghEcon],
                    "world fin crisis-gh":[cosine_worldFin],"Human body":[cosine_humanB] })
cosine_df = cosine_df.transpose()
cosine_df = cosine_df.rename(columns={0:"Cosine"})
cosine_df
```

# Term Frequencies

## Generating the Term Frequencies with the method;


```python
#Tranforming the Corpus vectors or presenting the Term Frequencies
corpus_vectors = cv.transform(corpus)
corpus_vectors = corpus_vectors.toarray()
corpus_vectors
```

```python
#Generating the Term Frequencues in the corpus into a dataframe df
df = pd.DataFrame(corpus_vectors, columns = cv.get_feature_names())
df
```

```python
#transposing the corpus to have documents as columns and terms as rows
df1 = df.transpose()
df1
```

```python
#renaming the columns to document names
tf = df1.rename(columns={0:"Query",1:"Ghana",2:"US",3:"Politics",4:"Chimera", 5:"Computer", 6:'math', 7:'enzyme',
                          8:'finance',
                         9:'Ghana Bank crisis',
                         10:'Ghana economic turmoil',
                         11:'world fin crisis-gh',
                         12:"Human body"})
tf.head(20)
```

```python
# Normalisation of the Term Frequencies

tf_norm = (tf-tf.min())/(tf.max() - tf.min())
tf_norm
```

## Cardinality of Terms and Proportionality Weights;


```python
#computing the cardinality for each document
card_dλ = tf.sum(axis=0)
card_dλ_tf = pd.DataFrame(card_dλ)
card_dλ = card_dλ_tf.transpose()
print('CARDINALITY')
card_dλ
# card_dλ.drop("Query", axis = 1)
```

## Conditional Cardinality of Terms and Proportionality Weights;


```python
# computing the conditional cardinality for each document
condition = tf['Query']>0
filtered_tf = tf.loc[condition]
Con_card_dλ_tf = filtered_tf[['Ghana', 'US','Politics','Chimera','Computer','math','enzyme','finance','Ghana Bank crisis', 'Ghana economic turmoil','world fin crisis-gh', 'Human body']].sum(axis=0)
Con_card_dλ_tf = pd.DataFrame(Con_card_dλ_tf)
Con_card_dλ = Con_card_dλ_tf.transpose()
print("CONDITIONAL CARD")
Con_card_dλ
```

## Computing the p value for each document


```python
#Computing the p values for each document

#p for Ghana
p_gh = Con_card_dλ.loc[0,'Ghana']/card_dλ.loc[0,'Ghana']
print("Ghana ->",p_gh)

#p for US
p_us = Con_card_dλ.loc[0,'US']/card_dλ.loc[0,'US']
print('US ->',p_us)

#p for Politics
p_politics = Con_card_dλ.loc[0,'Politics']/card_dλ.loc[0,'Politics']
print('Politics ->',p_politics)


#p for Chimera
p_Chimera = Con_card_dλ.loc[0,'Chimera']/card_dλ.loc[0,'Chimera']
print('Chimera ->',p_Chimera)

#p for Computer
p_Computer = Con_card_dλ.loc[0,'Computer']/card_dλ.loc[0,'Computer']
print('Computer ->',p_Computer)

#p for math
p_math = Con_card_dλ.loc[0,'math']/card_dλ.loc[0,'math']
print("Math ->",p_math)

#p for enzyme
p_enzyme = Con_card_dλ.loc[0,'enzyme']/card_dλ.loc[0,'enzyme']
print('Enzyme ->',p_enzyme)

#p for Finance
p_finance = Con_card_dλ.loc[0,'finance']/card_dλ.loc[0,'finance']
print('Finance ->',p_finance)

#p for Ghana Bank crisis
p_ghBank = Con_card_dλ.loc[0,'Ghana Bank crisis']/card_dλ.loc[0,'Ghana Bank crisis']
print('Ghana Bank crisis ->',p_ghBank)

#p for Ghana economic turmoil
p_ghEcon = Con_card_dλ.loc[0,'Ghana economic turmoil']/card_dλ.loc[0,'Ghana economic turmoil']
print('Ghana economic turmoil ->',p_ghEcon)


#p for world fin crisis-gh
p_worldFin = Con_card_dλ.loc[0,'world fin crisis-gh']/card_dλ.loc[0,'world fin crisis-gh']
print('World fin crisis-gh ->', p_worldFin)

#p for Human body
p_humanB = Con_card_dλ.loc[0,'Human body']/card_dλ.loc[0,'Human body']
print('Human body ->',p_humanB)
```

## Computation of the tf - p


```python
# computation of tf-p
dλ = tf_norm["Ghana"] * p_gh, tf_norm['US']*p_us, tf_norm['Politics']*p_politics,tf_norm['Chimera']*p_Chimera, tf_norm['Computer']*p_Computer, tf_norm['math']*p_math,tf_norm['enzyme']*p_enzyme, tf_norm['finance']*p_finance, tf_norm['Ghana Bank crisis']*p_ghBank, tf_norm['Ghana economic turmoil']*p_ghEcon, tf_norm['world fin crisis-gh']*p_worldFin, tf_norm['Human body']*p_humanB
dλ = pd.DataFrame(dλ)
dλ = dλ.transpose()
dλ
```
